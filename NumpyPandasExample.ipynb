```python
import pandas as pd
import numpy as np

# Create the same data directly
data = {
    'Name': ['John', 'Alice', 'Bob', 'Sara', 'Sam', 'Rohit'],
    'Age': [20, 22, 21, 23, 25, 23],
    'Marks': [85, 90, 78, 92, 90, 45]
}

# Convert to DataFrame
df=pd.DataFrame(data)

# Print the DataFrame
print(df)

# Convert the 'Marks' column to a NumPy array
marks=df['Marks'].to_numpy()

print('\nNumpy Array of Marks:')
print(marks)
```

```python
print('Highest Marks ',np.max(marks))
```

```python
print('Lowest Marks ',np.min(marks))
```

```python
print('Average Marks ',np.mean(marks))
```

```python
print('Location : Label Based')
print(df.loc[2,['Name','Marks']])
```

```python
print('Location : Index Based')
print(df.iloc[2,[1,2]])
```

# Find the student name who got highest marks

```python
top=df.loc[df['Marks'].idxmax(),'Name']
print('Top Student Name ',top)
```

```python
last=df.loc[df['Marks'].idxmin(),'Name']
print('Last Student Name ',last)
```

```python
top_index = df['Marks'].idxmax()
top_student = df.iloc[top_index,df.columns.get_loc('Name')]
print('The top student is',top_student)
     
```
