# Scatter Plots
# In scatter plots we represent data in points, where each point represents a data observation , we can check how the values correlate

```python
import matplotlib.pyplot as plt

product_sales = {'Laptop': 750, 'Smart Phone': 650, 'Smart Watch': 400, 'Tablets': 500, 'AC': 400}
products = list(product_sales.keys())
sales = list(product_sales.values())
plt.scatter(products,sales)
plt.title('Basic product and sales representation')
plt.xlabel('products')
plt.ylabel('sales')
plt.grid(True)
plt.show()
```

```python
players=['MSD','Rohit','Virat','Gill','Shami']
match1_score=[45,67,56,78,89]
match2_score=[55,78,40,80,97]
match3_score=[90,40,20,30,60]
plt.figure(figsize=(8,5))
plt.scatter(players,match1_score,color='skyblue',label='Match One',s=100)
plt.scatter(players,match2_score,color='orange',label='Match Two',s=100,marker='d')
plt.scatter(players,match3_score,color='green',label='Match Three',s=150,marker='s')
plt.title('Comparison of players in three matches')
plt.xlabel('players')
plt.ylabel('Score')
plt.legend(loc='lower left')
plt.tight_layout()
plt.grid('True',linestyle='--',alpha=0.2)
plt.show()
```

# Write a comparison that shows average calories burned for each workout type from final_data dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Final_data.csv')
avg_cal = df.groupby('Workout_Type')['Calories_Burned'].mean()
selected_index = avg_cal.index
selected_value = avg_cal.values

plt.scatter(selected_index, selected_value, color='skyblue', s=100)
plt.title('Comparison of Workout and Calories Burned')
plt.xlabel('Workout Type')
plt.ylabel('Calories Burned')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
