```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
temp = np.random.randint(70, 80, size=50)

plt.scatter(x, y, c=temp, cmap='plasma', s=120, edgecolors='black')
plt.colorbar(label='Temperature')
plt.title('Scatter plot Colored by Temperature')
plt.xlabel('X-Random')
plt.ylabel('Y-Random')
plt.show()
```

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,56,59,62,66,68,72]
age_year=[20,22,25,28,30,35,38,40]
plt.figure(figsize=(8,5))
plt.scatter(height_cm,weight_kg,c=age_year,cmap='rainbow',s=120,edgecolor='black')
plt.colorbar(label='Age (years)')
plt.title('Scatter plot Colored by Temperature')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.grid(color='skyblue',linestyle='--',linewidth=0.4)
plt.show()
```

# Pie Plot Example
    ## plt.pie(x,labels='',colors='',autoopct='',startangle='',shadow=,explode=,radius)

```python
depts=['HR','web development','app development','agile scrum','testing']
expenses=[30,10,30,15,15]
plt.figure(figsize=(7,7))
```

```python
plt.pie(expenses,labels=depts,autopct='%1.2f%%',startangle=140,explode=(0, 0.05, 0, 0, 0),shadow=True
)
plt.title('Department-wise Expense Report')
plt.show()
```

```python
plt.pie(expenses,labels=depts,colors=plt.cm.cool([0.1,0.3,0.5,0.7,0.9]),autopct='%1.2f%%',startangle=140,explode=(0, 0.05, 0, 0, 0),shadow=True
)
plt.title('Department-wise Expense Report')
plt.show()
```

```python
plt.pie(expenses,labels=depts,colors=plt.cm.inferno([0.6,0.5,0.9,0.4,0.8]),autopct='%1.0f%%',startangle=140,explode=(0, 0.05, 0, 0, 0),shadow=True
)
plt.title('Department-wise Expense Report')
plt.show()
```

# Create a pie plot 

```python
sales_data = {'Mobile': 45000, 'Clothing': 30000, 'Groceries': 35000, 'Home Decor': 20000}
labels = list(sales_data.keys())
values = list(sales_data.values())
colors = plt.cm.plasma(np.linspace(0, 1, len(labels)))

plt.figure(figsize=(7,7))
plt.pie(values, labels=labels, autopct='%1.2f%%', colors=colors,
    textprops={'color':'skyblue','fontsize':10,'fontweight':'bold'},
        startangle=90)
plt.title('Sales Distribution by Category')
plt.show()
```
