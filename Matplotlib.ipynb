# First Example to create a line chart from our data 
    # Students.xlsx for fetching the data
    # Generate Line Chart based on student name and student marks

```python
import pandas as pd
```

```python
import pandas as pd
import numpy as np

# Create dataset
data = {
    'Name': ['John', 'Alice', 'Bob', 'Sara', 'Sam', 'Rohit'],
    'Age': [20, 22, 21, 23, 25, 23],
    'Marks': [85, 90, 78, 92, 90, 45]
}

df = pd.DataFrame(data)
print(df)

```

```python
names=df['Name']
marks=df['Marks']
print(names)
print(marks)
```

```python
plt.plot(names,marks,marker='s',color='green')
plt.title('Student Marks Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(names,marks,marker='o',color='orange')
plt.title('Student Progress Chart')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(names,marks,marker='d',color='red',linestyle='--')
plt.title('Student Progress Chart')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```
