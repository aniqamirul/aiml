# 2. Logistic Regression - Pass/Fail

```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('pass_fail.csv')
df.head()
```

```python
X = df[['Hours']]
y = df['Result']
```

```python
model=LogisticRegression()
model.fit(X,y)
```

```python
pred_35 = model.predict([[3.5]])
print('Predicted result for 3.5 hours:', pred_35[0])
```

```python
hours_study=float(input('Number of hours studied'))
pred_value=model.predict([[hours_study]])
print(f'AFter {hours_study} hours study you are likely to {pred_value[0]}')
```

```python
new_cust_age = int(input('Enter Age: '))
new_cust_sal = int(input('Enter Salary: '))
new_cust = [[new_cust_age, new_cust_sal]]

prediction = model.predict(new_cust)

if prediction:
    print('Purchased')
else:
    print('Not Purchased')
```

# Logistic Regression — Spam/NotSpam
# Download spam_data.csv from https://www.kaggle.com/

# Do the data cleaning and labeling (if required)
# Use Logistic Regression Algorithm to trained the Model
# Predict Result if the email is spam or not spam

```python
# Step 1: Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Step 2: Load the Data
df = pd.read_csv('spam_dataset.csv')   # make sure the file is in the same folder
print(df.head())
```

```python
X = df['message_content']
y = df['is_spam']

# Step 5: Convert text to numbers
cv = CountVectorizer()
X_cv = cv.fit_transform(X)

# Step 6: Train model
model = LogisticRegression()
model.fit(X_cv, y)
```

```python
# Step 8: Try custom messages
sample = ["Win a free iPhone now!", "Are you coming to class today?"]
pred = model.predict(cv.transform(sample))

for msg, label in zip(sample, pred):
    print(msg, "->", "SPAM" if label == 1 else "NOT SPAM")
```
