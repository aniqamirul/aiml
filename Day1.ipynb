```python
course = "Python"

#  P   y   t   h   o   n
#  0   1   2   3   4   5   â†’ Positive / Forward Indexing
# -6  -5  -4  -3  -2  -1   â†’ Negative / Backward Indexing

# Python provides index numbers to each character
# Index number always starts from 0
# Indexing allows us to get only one single value
# Python uses [] notation for indexing

print(course)        # Prints the whole string
print(course[3])     # Indexing â†’ returns 'h'
print(course[0])     # Indexing â†’ returns 'P'
print(course[-5])    # Negative indexing â†’ returns 'y'
```
